```python
import pandas as pd


data = pd.read_csv("merged_data.csv")



```

```python
# Analyze total sales by customer segments
total_sales_by_customer = data.groupby('PREMIUM_CUSTOMER')['TOT_SALES'].sum()
print("Total Sales by Customer Segment:")
print(total_sales_by_customer)
```

```python
# Calculate the total sales for each premium customer category
total_sales_by_customer = data.groupby('PREMIUM_CUSTOMER')['TOT_SALES'].sum().reset_index()

# Plot the grouped bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='PREMIUM_CUSTOMER', y='TOT_SALES', data=total_sales_by_customer)
plt.xlabel('Premium Customer')
plt.ylabel('Total Sales')
plt.title('Total Sales by Premium Customer')
plt.show()

```

```python

# Analyze total sales by lifestage
total_sales_by_lifestage = data.groupby('LIFESTAGE')['TOT_SALES'].sum()
print("Total Sales by Lifestage:")
print(total_sales_by_lifestage)
```

```python
# Calculate the total sales for each premium customer category

# Calculate total sales by lifestage
total_sales_by_lifestage = data.groupby('LIFESTAGE')['TOT_SALES'].sum().reset_index()

# Plot the grouped bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='LIFESTAGE', y='TOT_SALES', data=total_sales_by_lifestage)
plt.xlabel('Lifestage')
plt.ylabel('Total Sales')
plt.title('Total Sales by Lifestage')
plt.xticks(rotation=45)
plt.show()


```

```python
# Analyze average sales per transaction
average_sales_per_transaction = data['TOT_SALES'].mean()
print("Average Sales per Transaction:", average_sales_per_transaction)

```

```python
total_sales_by_product = data.groupby('PROD_NAME')['TOT_SALES'].sum().sort_values(ascending=False)
print("Total Sales by Product Name:")
print(total_sales_by_product)




```

```python
total_sales_by_product = data.groupby('PROD_NAME')['TOT_SALES'].sum()

# Get the product name with the highest total sales
top_product_name = total_sales_by_product.idxmax()

# Get the corresponding total sales value
top_total_sales = total_sales_by_product.max()

print("Product with the highest total sales:", top_product_name)
print("Total sales for this product:", top_total_sales)
```

```python
total_sales_by_product = data.groupby('PROD_NAME')['TOT_SALES'].sum().reset_index()
top_10_products = total_sales_by_product.nlargest(10, 'TOT_SALES')

plt.figure(figsize=(10, 12))
sns.barplot(x='TOT_SALES', y='PROD_NAME',data=top_10_products, orient='h')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.title('Total Sales by Product')
plt.show()
```

```python
# Group the data by 'PROD_NAME' and sum the 'PROD_QTY' for each product
product_quantity = data.groupby('PROD_NAME')['PROD_QTY'].sum()

# Find the product with the highest total quantity sold
most_sold_product = product_quantity.idxmax()
total_quantity_sold = product_quantity.max()

print("Product with the most quantity sold:", most_sold_product)
print("Total quantity sold:", total_quantity_sold)
```

```python
# Group the data by 'PROD_NAME' and sum the 'PROD_QTY' for each product
product_quantity = data.groupby('PROD_NAME')['PROD_QTY'].sum()

# Find the product with the highest total quantity sold
mostless_sold_product = product_quantity.idxmin()
total_quantity_sold = product_quantity.min()

print("Product with the min quantity sold:", mostless_sold_product)
print("Total quantity sold:", total_quantity_sold)


```
